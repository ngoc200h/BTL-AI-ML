```python
data.drop(['review_id', 'user_name','thumbs_up', 'review_title', 'developer_response', 'developer_response_date', 'appVersion'], axis=1, inplace=True)
```

```python
# lower case
data.loc[:,'review_description'] = data.loc[:,'review_description'].apply(lambda x: " ".join(x.lower() for x in x.split()))
data.head(20)
```

```python
# remove punctuation
```

```python
import re
data.loc[:,'review_description'] = data.loc[:,'review_description'].apply(lambda word: re.sub('[^a-z A-Z 0-9-]+', '', word))
data.head(10)
```

```python
# remove stopwords
```

```python
import nltk
from nltk.corpus import stopwords

# Download the 'stopwords' resource
nltk.download('stopwords')

# Define the list of stopwords
stop_words = stopwords.words('english')
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
data.loc[:,'review_description'] = data.loc[:,'review_description'].apply(lambda word: " ".join(word for word in word.split() if word not in stop_words))
data.head(10)
```

```python
lemmat = WordNetLemmatizer()
data.loc[:,'review_description'] = data.loc[:,'review_description'].apply(lambda word: " ".join(lemmat.lemmatize(word) for word in word.split()))
data.head(10)
```

```python
def remove_urls(review_description):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', review_description)
```

```python
data['review_description'] = data['review_description'].apply(remove_urls)
```

```python
def remove_html(review_description):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', review_description)
```

```python
data['review_description'] = data['review_description'].apply(remove_html)
```

```python
chat_words_str = """
asap=as soon as possible
atm=at the moment
bbl=be back later
bbs=be back soon
brb=be right back
brt=be right there
btw=by the way
b4=before
faq=frequently asked questions
fyi=for your information
gal=get a life
gg=good game
gmta=great minds think alike
gr8=great!
g9=genius
ic=i see
imho=in my honest/humble opinion
imo=in my opinion
iow=in other words
irl=in real life
lol=laughing out loud
l8r=later
mte=my thoughts exactly
nrn=no reply necessary
oic=oh i see
prt=party
thx=thank you
"""
```

```python
chat_words_map_dict = {}
chat_words_list = []

for line in chat_words_str.split("\n"):
    if line != "":
        cw = line.split("=")[0]
        cw_expanded = line.split("=")[1]
        chat_words_list.append(cw)
        chat_words_map_dict[cw] = cw_expanded

chat_words_list = set(chat_words_list)

def chat_words_conversion(review_description):
    new_review_description = []
    for word in review_description.split():
        if word.lower() in chat_words_list:
            new_review_description.append(chat_words_map_dict[word.lower()])
        else:
            new_review_description.append(word)
    return " ".join(new_review_description)

data['reviews_description'] = data['review_description'].apply(chat_words_conversion)
```

```python
# Remove rows with empty or spaces-only 'review_description'
data = data[data['review_description'].str.strip() != '']
```
